```python
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.5)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
import numpy as np
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.metrics import accuracy_score
```

```python
accu = []
pred = []
```

```python
from sklearn.metrics import confusion_matrix
```

```python
#Utilización de KNN para entrenar el modelo
for w in ['uniform','distance']:
    for i in range(1,50):
        knn = KNeighborsClassifier(i, w)
        knn.fit(X_train, y_train)
        p = knn.predict(X_test)
        pred.append(p)
        accu.append(accuracy_score(y_test, p))
print ("El mejor modelo es para k = ", np.argmax(accu), ", con score", accu[np.argmax(accu)])
print ("El peor modelo es para k = ", np.argmin(accu), ", con score", accu[np.argmin(accu)])
```

```python
#Matrices de confusión para el mejor y el peor modelo
print( "MATRIZ DE CONFUSION")
print("Para el MEJOR MODELO")
print(confusion_matrix(y_test,pred[np.argmax(accu)]))
print("Para el PEOR MODELO")
print(confusion_matrix(y_test, pred[np.argmin(accu)]))
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn import datasets
from sklearn import svm
```

```python
iris = datasets.load_iris()
```

```python
iris.data.shape, iris.target.shape
```

```python
 X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.5, random_state=0)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
accu = []
clfarray = []
```

```python
#Cross validation
for i in range(1,50):
    clf = svm.SVC(i)
    clfarray.append(clf)
    accu.append(np.mean(cross_val_score(clf, iris.data, iris.target, cv=5)))
print ("El mejor modelo es para k = ", np.argmax(accu), ", con score", accu[np.argmax(accu)])
print ("El peor modelo es para k = ", np.argmin(accu), ", con score", accu[np.argmin(accu)])
```

```python
from sklearn.externals import joblib
```

```python
#Guardamos el mejor modelo para usos futuros
joblib.dump(clfarray[np.argmax(accu)], 'mejor_modelo.pkl', protocol=2)
```

```python
#SVM
for i in range(1,50):
    clf = svm.SVC(i)
    clf.fit(X_train, y_train)
    p = clf.predict(X_test)
    accu.append(accuracy_score(y_test, p))
print ("El mejor modelo es para k = ", np.argmax(accu), ", con score", accu[np.argmax(accu)])
print ("El peor modelo es para k = ", np.argmin(accu), ", con score", accu[np.argmin(accu)])
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
accu = []
```

```python
#Random Forest
for i in range(1,50):
    clf = RandomForestClassifier(n_estimators=10)
    clf = clf.fit(X_train, y_train)
    accu.append(clf.score(X_train, y_train))
print ("El mejor modelo es para k = ", np.argmax(accu), ", con score", accu[np.argmax(accu)])
print ("El peor modelo es para k = ", np.argmin(accu), ", con score", accu[np.argmin(accu)])
```
