```python
import sklearn.preprocessing as preprocessing
```

```python
#Para poder hacer gráficas correctamente, codificamos las variables categóricas ("race", "education",...) como variables
#numéricas.
def number_encode_features(df):
    result = df.copy()
    encoders = {}
    for column in result.columns:
        if result.dtypes[column] == np.object:
            encoders[column] = preprocessing.LabelEncoder()
            result[column] = encoders[column].fit_transform(result[column])
    return result, encoders

#Calculamos la correlación de las columnas y dibujamos un heatmap, que nos mostrará las variables correlacionadas.
encoded_data, _ = number_encode_features(adult_modified)
sns.heatmap(encoded_data.corr(), square=True)
plt.show()
```

```python
#Según el mapa de color, podemos ver que entre "education" y "education-num" y entre "marital-status" y "relationship" 
#hay una alta correlación, y entre "relationship" y "sex" hay una muy baja.
#Hacemos una gráfica de la relación entre "education" y "education-num", para ver qué es lo que representa
#"education-num"
sns.stripplot(x="education-num", y="education", data=adult_modified)
```

```python
#Como podemos ver, cada "education" tiene asociada un "education-num", por lo que podemos suponer que cada "education-num"
#representa un "education".
#Es interesante ver también que, por ejemplo, el doctorado ("Doctorate") tiene el "education-num" más alto y 
#el preescolar ("Preschool") tiene el más bajo, osea que la columna "education-num" va ordenada por nivel de educación.
adult_modified[["education", "education-num"]].head(20)
```

```python
from sklearn.model_selection import train_test_split
```

```python
encoded_data.ix[:,encoded_data.columns != "education-num"]
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import cross_val_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(encoded_data.ix[:,encoded_data.columns != "target"], 
                                                    encoded_data["target"], test_size=0.5)
```

```python
#Empezamos con la predicción
#Random Forest
clf = RandomForestClassifier()
clf = clf.fit(X_train, y_train)
score = cross_val_score(clf, X_test, y_test, cv=5)
score.mean()
```

```python
#Pasamos al tuneo de variables para el random forest
#Utilizamos GridSearch para encontrar los valores más adecuados
from sklearn.grid_search import GridSearchCV

# Buscamos buenos valores para los hiperparámetros
# 
n_estimators = list(np.arange(200, 1100, 400)) #arange(start,stop,step)
max_features = [None, 'sqrt', 'log2'] 

hyperparameters = {
    'n_estimators': n_estimators, 
    'max_features': max_features
}

# Grid search using cross-validation
gridCV = GridSearchCV(RandomForestClassifier(), param_grid=hyperparameters, cv=5)
```

```python
# Perform grid search with 5-fold CV
gridCV.fit(X_train, y_train)

# Identify optimal hyperparameter values
best_n_estim      = gridCV.best_params_['n_estimators']
best_max_features = gridCV.best_params_['max_features']  

print("The best performing n_estimators value is: {}".format(best_n_estim))
print("The best performing max_features value is: {}".format(best_max_features))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# fetch scores, reshape into a grid
scores = [x[1] for x in gridCV.grid_scores_]
scores = np.array(scores).reshape(len(n_estimators), len(max_features))
scores = np.transpose(scores)

# Make heatmap from grid search results
plt.figure(figsize=(12, 6))
plt.imshow(scores, interpolation='nearest', origin='higher', cmap='jet_r')
plt.xticks(np.arange(len(n_estimators)), n_estimators)
plt.yticks(np.arange(len(max_features)), max_features)
plt.xlabel('Number of decision trees')
plt.ylabel('Max features')
plt.colorbar().set_label('Classification Accuracy', rotation=270, labelpad=20)
plt.show()
```

```python
#Probamos con los valores dados por el mapa de calor
clf = RandomForestClassifier(n_estimators=1000, max_features="sqrt")
clf = clf.fit(X_train, y_train)
score = cross_val_score(clf, X_test, y_test, cv=5)
score.mean()
```

```python
#Podemos ver que el accuracy mejora un poco, se va aproximando al 85%
#Ahora probamos con hiperparámetros manualmente
clf = RandomForestClassifier(n_estimators=200, max_features="log2", min_samples_split=10)
clf = clf.fit(X_train, y_train)
score = cross_val_score(clf, X_test, y_test, cv=5)
score.mean()
```

```python
#Vemos que el score supera el 85%
#Seguimos con el tuneo
clf = RandomForestClassifier(n_estimators=1000, max_features="log2", min_samples_split=10)
clf = clf.fit(X_train, y_train)
score = cross_val_score(clf, X_test, y_test, cv=5)
score.mean()
```

```python
clf = RandomForestClassifier(n_estimators=600, max_features="log2", min_samples_split=10, min_samples_leaf=5)
clf = clf.fit(X_train, y_train)
score = cross_val_score(clf, X_test, y_test, cv=5)
score.mean()
```
